```python
# Matplotlib - all properties - scatter plot

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("cleaned_data.csv")

prices = df["price"].astype(int)
price_counts=df.groupby("price").size()

plt.figure()
plt.scatter(price_counts.index, price_counts.values, alpha=0.6)

plt.title("Price Distribution ()")
plt.xlabel("Price ()")
plt.ylabel("Number of properties")
```

```python
# Matplotlib - all properties - histogram
plt.hist(df["price"], bins=10)
plt.title("Distribution of Property Prices")
plt.xlabel("Price ( millions)")
plt.ylabel("Number of Listings")
plt.show()
```

```python
# Seaborn - all properties - histogram

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("cleaned_data.csv")
prices = df["price"].astype(int)

sns.histplot(data=df,x="price", bins=30, kde=True)
plt.xlabel("Price (m)")
plt.ylabel("# of Properties")
plt.title("Distribution of Property Prices")

plt.show
```

```python
# Seaborn - Houses - histogram

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("cleaned_data.csv")
print(df.columns.tolist())

prices = df["price"].astype(int)
house_df=df[df["type"].str.contains("House", case=False, na=False)]

print(f"Total listings: {len(df):,}")
print(f"House listings: {len(house_df):,}")
print(house_df["subtype"].value_counts())

plt.figure()
sns.histplot(data=house_df,x="price", bins=30, kde=True)
plt.xlabel("Price (m)")
plt.ylabel("# of Properties")
plt.title("Distribution of House Prices")

plt.show
```

```python
# Seaborn - Houses IXELLES - histogram

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("cleaned_data.csv")
print(df.columns.tolist())

df["price"] = df["price"].astype(int)

house_df=df[df["type"].str.contains("House", case=False, na=False)]
house_1050 = house_df[house_df["postCode"].astype(str).str.startswith("1050")]
house_SubtypeHouse= house_1050[house_1050["subtype"].str.contains("House", case=False, na=False)]

print(f"Total listings: {len(df):,}")
print(f"House listings: {len(house_SubtypeHouse):,}")
print(house_SubtypeHouse["subtype"].value_counts())

plt.figure(figsize=(8,5))
sns.histplot(data=house_SubtypeHouse,x="price", bins=30, kde=True)
plt.xlabel("Price (m)")
plt.ylabel("# of Properties")
plt.title("Distribution of House Prices in IXELLES 1050 SubType House")

plt.show
```
